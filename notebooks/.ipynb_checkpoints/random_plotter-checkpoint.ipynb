```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-pastel')
# plt.style.use('seaborn-whitegrid')
import numpy as np
from sklearn.preprocessing import StandardScaler, scale
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x), color='cyan')
ax.plot(x, np.cos(x), color='y');
```

```python
# def normalize(v):
#     norm = np.linalg.norm(v)
#     if norm == 0: 
#         return v
#     return v / norm
def normalize(v):
#     return (v - v[0])
    return scale(np.array(v), with_mean=True, with_std=False)
#     ss = StandardScaler()
#     return ss.fit_transform(np.array(v))
```

```python
fig = plt.figure(figsize=(8, 20))
x = [2016, 2017, 2018]

ax1 = fig.add_subplot(3, 1, 1)
y_lasso = np.array([3.6411, 3.7305, 3.7452])
y_lgbm = np.array([3.2637, 3.4153, 3.4608])
y_lstm = np.array([2.9523, 3.0971, 3.0971])
ax1.plot(x, normalize(y_lasso), '-ok', color='g', label='Lasso')
ax1.plot(x, normalize(y_lgbm), '-ok', color='b', label='LightGBM')
ax1.plot(x, normalize(y_lstm), '-ok', color='r', label='LSTM')
ax1.legend(loc='upper left')
ax1.xaxis.set_major_locator(plt.FixedLocator(x))
ax1.set(xlabel='Prediction Year', 
       ylabel='Mean Absolute Error (Standard Scale)', 
       title='Inference When Trained on 1 Previous Year')

ax2 = fig.add_subplot(3, 1, 2)
y_lasso = np.array([3.2905, 3.2960, 3.3335])
y_lgbm = np.array([2.9430, 3.0060, 3.0517])
y_lstm = np.array([2.6470, 2.6662, 2.6925])
ax2.plot(x, normalize(y_lasso), '-ok', color='g', label='Lasso')
ax2.plot(x, normalize(y_lgbm), '-ok', color='b', label='LightGBM')
ax2.plot(x, normalize(y_lstm), '-ok', color='r', label='LSTM')
ax2.legend(loc='upper left')
ax2.xaxis.set_major_locator(plt.FixedLocator(x))
ax2.set(xlabel='Prediction Year', 
       ylabel='Mean Absolute Error (Standard Scale)',
       title='Inference When Trained on 2 Previous Years')


ax3 = fig.add_subplot(3, 1, 3)
y_lasso = np.array([3.1154, 3.1080, 3.1156])
y_lgbm = np.array([2.7762, 2.8294, 2.8815])
y_lstm = np.array([2.4836, 2.4890, 2.5067])
ax3.plot(x, normalize(y_lasso), '-ok', color='g', label='Lasso')
ax3.plot(x, normalize(y_lgbm), '-ok', color='b', label='LightGBM')
ax3.plot(x, normalize(y_lstm), '-ok', color='r', label='LSTM')
ax3.legend(loc='upper left')
ax3.xaxis.set_major_locator(plt.FixedLocator(x))
ax3.set(xlabel='Prediction Year', 
       ylabel='Mean Absolute Error (Standard Scale)',
       title='Inference When Trained on 3 Previous Years');
```
